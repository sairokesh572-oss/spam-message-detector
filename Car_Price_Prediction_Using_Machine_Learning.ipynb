```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([500, 420, 360, 300, 260])

# Polynomial transformation
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, y)

# Predict for a specific age
age = np.array([[3.5]])
age_poly = poly.transform(age)
price = model.predict(age_poly)
print("Predicted Car Price:", price[0])

# Generate smooth curve for visualization
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_range_pred = model.predict(X_range_poly)

# Plotting
plt.scatter(X, y, color='blue', label='Actual Data', s=60)
plt.plot(X_range, y_range_pred, color='red', label='Polynomial Fit (Degree 2)', linewidth=2)
plt.scatter(age, price, color='green', marker='o', s=100, label=f'Predicted (Age=3.5, â‚¹{price[0]:.0f})')

plt.title('Polynomial Regression - Car Price vs Age')
plt.xlabel('Car Age (years)')
plt.ylabel('Car Price')
plt.legend()
plt.grid(True)
plt.show()
```
